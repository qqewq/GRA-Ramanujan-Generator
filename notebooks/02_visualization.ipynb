## 1. Настройка окружения

```python
import sys
sys.path.append('..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from gra_ramanujan.visualization.plots import plot_evolution, plot_heatmap, plot_complexity_distribution
from gra_ramanujan.visualization.gallery import generate_latex_gallery
```

## 2. Загрузка логов

Результаты эволюции сохраняются в файлах `data/conjectures.csv` (все сгенерированные формулы) и `data/machines.csv` (статистика по агентам).

```python
df_conj = pd.read_csv('../data/conjectures.csv')
df_mach = pd.read_csv('../data/machines.csv')

print(f"conjectures.csv: {len(df_conj)} записей")
print(f"machines.csv: {len(df_mach)} записей")
df_conj.head()
```

## 3. Эволюция красоты формул

Функция `plot_evolution()` строит график максимальной красоты формул для каждой константы по поколениям.

```python
plot_evolution()
```

На этом графике видно, как с течением времени лучшие формулы становятся красивее (растёт точность и падает сложность). Разные линии соответствуют разным константам – можно сравнить, для каких констант поиск идёт успешнее.

## 4. Теплокарта прогресса

`plot_heatmap()` создаёт теплокарту, где по оси X – поколения, по оси Y – константы, а цвет показывает максимальную красоту формулы для данной константы в данном поколении.

```python
plot_heatmap()
```

Тёмные участки означают, что для этой константы найдены красивые формулы. Можно заметить, какие константы «даются» эволюции легче.

## 5. Распределение сложности

`plot_complexity_distribution()` показывает ящики с усами (boxplot) распределения сложности формул по поколениям.

```python
plot_complexity_distribution()
```

Обычно на ранних поколениях формулы могут быть сложнее, а затем популяция отбирает более простые и точные выражения.

## 6. Динамика витальности агентов

Построим график средней витальности популяции и отдельных агентов.

```python
plt.figure(figsize=(10,5))
# Средняя витальность по поколениям
mean_vitality = df_mach.groupby('generation')['vitality'].mean()
plt.plot(mean_vitality.index, mean_vitality.values, 'o-', label='Средняя витальность')

# Отобразим несколько случайных агентов
sample_ids = np.random.choice(df_mach['machine_id'].unique(), min(5, len(df_mach['machine_id'].unique())), replace=False)
for mid in sample_ids:
    data = df_mach[df_mach['machine_id'] == mid]
    plt.plot(data['generation'], data['vitality'], '--', alpha=0.5, label=f'Agent {mid}')

plt.xlabel('Поколение')
plt.ylabel('Витальность')
plt.title('Динамика витальности агентов')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Видно, что агенты с высокой витальностью выживают, а низкая витальность ведёт к замене.
