# GRA-Ramanujan-Generator: Введение

Этот ноутбук знакомит с проектом **GRA-Ramanujan-Generator** – эволюционным синтезом математических формул на основе многоуровневой GRA-обнулёнки.

## 1. Концепция

Мы создаём популяцию ИИ-агентов («машин Рамануджана»), которые генерируют формулы для математических констант (π, e, γ, ζ(2) и др.). Каждый агент живёт в иерархии целей:

* **Уровень 0** – точность и простота формулы для конкретной константы.
* **Уровень 1** – согласованность формул для разных констант (известные тождества).
* **Уровень 2** – разнообразие и кооперация в популяции.

Агенты обладают витальностью: удачные находки увеличивают запас жизни, ошибки – уменьшают. Естественный отбор ведёт к появлению «гениальных» формул.

## 2. Настройка окружения

Убедитесь, что вы установили зависимости из `requirements.txt` и находитесь в корневой папке проекта. Для импорта модулей из `gra_ramanujan` добавим путь:

```python
import sys
sys.path.append('..')

import numpy as np
import matplotlib.pyplot as plt
from gra_ramanujan.universe import GRAUniverse
from gra_ramanujan.constants import CONSTANTS
from gra_ramanujan.utils import init_logs
from gra_ramanujan.visualization.plots import plot_evolution, plot_heatmap
```

## 3. Запуск эволюции с одной константой

Для начала ограничимся только константой π. Посмотрим, как агенты будут искать представления для числа π.

```python
# Инициализация логов
init_logs()

config = {
    'lambda0': 1.0,
    'lambda1': 0.5,
    'lambda2': 0.2,
    'death_threshold': 0.2,
    'mutation_rate': 0.1,
    'max_depth': 5,
    'const_range': (-10, 10),
    'beauty_alpha': 0.5,
    'share_prob': 0.1
}

# Берём только первую константу (π)
single_const = [CONSTANTS[0]]
universe = GRAUniverse(num_agents=10, config=config, constants=single_const)

print("Запуск эволюции на 30 поколений...")
for gen in range(30):
    universe.step()
    if gen % 5 == 0:
        print(f"Поколение {gen:2d}, J = {universe.history[-1]:.6f}, агентов: {len(universe.agents)}")
```

## 4. Визуализация

Построим график эволюции максимальной красоты формул.

```python
plot_evolution()
```

## 5. Лучшая найденная формула

Посмотрим в лог `data/conjectures.csv`, какая формула для π оказалась самой красивой.

```python
import pandas as pd
df = pd.read_csv('../data/conjectures.csv')
best_pi = df[df['constant'] == 'pi'].sort_values('beauty', ascending=False).iloc[0]
print(f"Лучшая формула для π:")
print(f"   Выражение: {best_pi['expression']}")
print(f"   Точность: {best_pi['accuracy']:.6f}")
print(f"   Сложность: {best_pi['complexity']}")
print(f"   Красота: {best_pi['beauty']:.6f}")
```

## 6. Запуск с несколькими константами

Теперь попробуем добавить несколько констант (π, e, γ) и посмотрим на теплокарту прогресса.

```python
multi_const = [c for c in CONSTANTS if c.name in ('pi', 'e', 'gamma')]
universe2 = GRAUniverse(num_agents=15, config=config, constants=multi_const)

for gen in range(40):
    universe2.step()
    if gen % 10 == 0:
        print(f"Поколение {gen:2d}, J = {universe2.history[-1]:.6f}")

plot_heatmap()
```

Теплокарта показывает, для каких констант эволюция идёт быстрее. Более тёмные клетки соответствуют более красивым формулам.
