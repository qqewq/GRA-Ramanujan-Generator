# Быстрый старт с GRA-Ramanujan-Generator
В этом туториале мы запустим простую эволюцию с одной константой и посмотрим на результаты.

```python
import sys
sys.path.append('..')
from gra_ramanujan.universe import GRAUniverse
from gra_ramanujan.constants import CONSTANTS
from gra_ramanujan.utils import init_logs
from gra_ramanujan.visualization.plots import plot_evolution

init_logs()
config = {
    'lambda0': 1.0,
    'lambda1': 0.5,
    'lambda2': 0.2,
    'death_threshold': 0.2,
    'mutation_rate': 0.1,
    'max_depth': 4,
    'const_range': (-5,5),
    'beauty_alpha': 0.5
}
universe = GRAUniverse(num_agents=5, config=config, constants=CONSTANTS[:1])  # только pi

for gen in range(20):
    universe.step()
    if gen % 5 == 0:
        print(f'Generation {gen}, J = {universe.history[-1]:.4f}')

plot_evolution()
```

## Анализ
На графике видно, как максимальная красота формул растёт со временем. Лучшая формула будет записана в лог.
